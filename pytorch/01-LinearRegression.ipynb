```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from torch.utils.tensorboard import SummaryWriter

x_train = torch.FloatTensor([1,2,3]).unsqueeze(1)
y_train = torch.FloatTensor([2,4,6]).unsqueeze(1)
W = torch.zeros(1, requires_grad=True)
b = torch.zeros(1, requires_grad=True)
optimizer = torch.optim.ASGD([W,b], lr=0.01)
```

```python
nb_epochs = 1000
epochs = np.arange(1, nb_epochs+1)
y_cost = []
y_W = []
y_b = []

for epoch in epochs:
    optimizer.zero_grad()
    # Forward Prop
    hypothesis = x_train * W + b
    cost = torch.mean((hypothesis - y_train)**2)
    # Backward Prop
    cost.backward()
    optimizer.step()
    # for plotting
    y_cost.append(float(cost))
    y_W.append(float(W))
    y_b.append(float(b))
```

```python
# 각 주기당 값들의 변화
def plot():
    fig, (ax1, ax2, ax3) = plt.subplots(3, 1)

    ax1.plot(epochs, y_cost)
    ax1.set_ylabel("cost")
    ax2.plot(epochs, y_W)
    ax2.set_ylabel("W")
    ax3.plot(epochs, y_b)
    ax3.set_ylabel("b")
    ax3.set_xlabel("epochs")

    fig.set_size_inches(12,10)
    plt.show()
    
plot()
```

```python
x_train = torch.FloatTensor([1,2,3]).unsqueeze(1)
y_train = torch.FloatTensor([2,4,6]).unsqueeze(1)
W = torch.zeros(1, requires_grad=True)
b = torch.zeros(1, requires_grad=True)
optimizer = torch.optim.ASGD([W,b], lr=0.01)
loss = torch.nn.MSELoss()

nb_epochs = 1000
epochs = np.arange(1, nb_epochs+1)
y_cost = []
y_W = []
y_b = []

for epoch in epochs:
    optimizer.zero_grad()
    # Forward Prop
    hypothesis = x_train * W + b
    cost = loss(hypothesis, y_train) # 이 줄만 바뀜
    # Backward Prop
    cost.backward()
    optimizer.step()
    # for plotting
    y_cost.append(float(cost))
    y_W.append(float(W))
    y_b.append(float(b))

plot()
```
